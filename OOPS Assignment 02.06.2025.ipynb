```python
# 8. Create a Python class called `BankAccount` with private attributes for the account balance (`__balance`) and account number (`__account_number`). Provide methods for depositing and withdrawing money.
class BankAccount:
    def __init__(self,account_no,_balance):
        self.__account_number = account_no
        self.__balance = _balance
    def deposit(self,amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ₹{amount}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ₹{amount}")
        else:
            print("Insufficient balance or invalid amount")

    def get_balance(self):
        return self.__balance
acc = BankAccount("1234567890", 5000)
acc.deposit(1000)
acc.withdraw(2000)
print("Remaining Balance:", acc.get_balance())
        

# 9. Discuss the advantages of encapsulation in terms of code maintainability and security.
'''advantages:
1.Data Security
2.controlled access:we decide who can access or modify the code
3.easy maintenance
4.code reusability
5.better debugging'''

# 10. How can you access private attributes in Python? Provide an example demonstrating the use of name mangling.
'''This is achieved by prefixing the attribute name with double underscores (__). When an attribute name starts with __, Python internally modifies 
the name by adding _ClassName before the original name. 
This is done to prevent accidental access or modification of these attributes from outside the class.  '''
class Demo:
    def __init__(self):
        self.__secret = "Hidden"

t = Demo()      
print(t._Demo__secret)     

```

```python
# 4. Discuss the benefits of using composition over inheritance in Python, especially in terms of code flexibility
# and reusability.
'''Code Flexibility:
we can change or replace components easily without changing the main class.
Reusability:
we can reuse components (objects of other classes) across different classes.
'''
# 5. How can you implement composition in Python classes? Provide examples of using composition to create
# complex objects.
'we can implement composition by creating objects of one class inside another class.'
class Processor:
    def start(self):
        print("Processor is starting...")

class Computer:
    def __init__(self):
        self.cpu = Processor()  

    def power_on(self):
        print("Computer is powering on...")
        self.cpu.start()
pc = Computer()
pc.power_on()
print("_____________________")

# 6. Create a Python class hierarchy for a music player system, using composition to represent playlists and
# songs.

class Song:
    def __init__(self, title, artist):
        self.title = title
        self.artist = artist

    def show(self):
        return f"{self.title} by {self.artist}"

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = [] 
        
    def add_song(self, song):
        self.songs.append(song)

    def show_songs(self):
        print(f"Playlist: {self.name}")
        for s in self.songs:
            print(" -", s.show())

class MusicPlayer:
    def __init__(self):
        self.playlists = [] 

    def add_playlist(self, playlist):
        self.playlists.append(playlist)

    def show_all(self):
        for p in self.playlists:
            p.show_songs()
            
song1 = Song("Blinding Light", "The Weeknd")
song2 = Song("Perfect", "Ed Sheeran")

playlist = Playlist("My Favorites")
playlist.add_song(song1)
playlist.add_song(song2)

player = MusicPlayer()
player.add_playlist(playlist)

player.show_all()

```

```python
# 7. Explain the concept of "has-a" relationships in composition and how it helps design software systems.
'''In composition, a class contains objects of other classes.
This is called a "has-a" relationship.
it is useful in software design:
1.Makes code modular (break into smaller parts)
2.Increases reusability (use same part in other classes)
3.Makes big systems easier to manage
4.Avoids long inheritance chains

'''
# 8. Create a Python class for a computer system, using composition to represent components like CPU, RAM,
# and storage devices.
class CPU:
    def process(self):
        print("CPU is processing data...")

class RAM:
    def load(self):
        print("RAM is loading programs...")

class Storage:
    def save(self):
        print("Storage is saving files...")

class Computer:
    def __init__(self):
        self.cpu = CPU()
        self.ram = RAM()
        self.storage = Storage()

    def start_computer(self):
        print("Starting computer...")
        self.cpu.process()
        self.ram.load()
        self.storage.save()
my_pc = Computer()
my_pc.start_computer()

# 9. Describe the concept of "delegation" in composition and how it simplifies the design of complex systems.
'''When a class passes a task to another class's object inside it.
Delegation is Helpful:
Keeps classes small and focused
Allows each class to do one job well
Makes code easy to update (just change one part)
Reduces duplication (reusing the same part in many places)
'''
```

```python
# 10. Create a Python class for a car, using composition to represent components like the engine, wheels, and transmission.
class Engine:
    def start(self):
        print("Engine started.")

class Wheels:
    def rotate(self):
        print("Wheels are rotating.")

class Transmission:
    def shift(self):
        print("Transmission is shifting gears.")

class Car:
    def __init__(self):
        self.engine = Engine()
        self.wheels = Wheels()
        self.transmission = Transmission()

    def drive(self):
        print("Car is starting to drive...")
        self.engine.start()
        self.transmission.shift()
        self.wheels.rotate()
my_car = Car()
my_car.drive()


# 11. How can you encapsulate and hide the details of composed objects in Python classes to maintain abstraction?
'''To hide internal parts (composed objects), you:
Make attributes private or protected using _ or __.
Only allow access through public methods.'''

# 12. Create a Python class for a university course, using composition to represent students, instructors, and course materials.

class Student:
    def __init__(self, name):
        self.name = name

class Instructor:
    def __init__(self, name):
        self.name = name

class CourseMaterial:
    def __init__(self, title):
        self.title = title

class Course:
    def __init__(self, course_name, instructor):
        self.course_name = course_name
        self.__instructor = instructor         
        self.__students = []                  
        self.__materials = []                

    def add_student(self, student):
        self.__students.append(student)

    def add_material(self, material):
        self.__materials.append(material)

    def course_details(self):
        print(f"Course: {self.course_name}")
        print(f"Instructor: {self.__instructor.name}")
        print("Students Enrolled:")
        for student in self.__students:
            print("-", student.name)
        print("Materials:")
        for mat in self.__materials:
            print("-", mat.title)

instructor = Instructor("Dr. A. Sharma")
course = Course("Python Programming", instructor)

course.add_student(Student("Riya"))
course.add_student(Student("Aman"))

course.add_material(CourseMaterial("OOP Notes"))
course.add_material(CourseMaterial("Assignment 1"))

course.course_details()

```

```python
# 16. Create a Python class for a computer game character, using composition to represent attributes like
# weapons, armor, and inventory.

class Weapon:
    def __init__(self, name):
        self.name = name

class Armor:
    def __init__(self, type_):
        self.type = type_

class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

class GameCharacter:
    def __init__(self, name, weapon, armor):
        self.name = name
        self.weapon = weapon    
        self.armor = armor      
        self.inventory = Inventory()  

sword = Weapon("Sword")
shield = Armor("Shield")
hero = GameCharacter("Archer", sword, shield)
hero.inventory.add_item("Health Potion")

print(f"Character: {hero.name}")
print(f"Weapon: {hero.weapon.name}")
print(f"Armor: {hero.armor.type}")
print(f"Inventory: {hero.inventory.items}")

# 17. Describe the concept of "aggregation" in composition and how it differs from simple composition.
'''Composition:
Strong "has-a" relationship.
The contained object cannot exist independently without the container.
If the container is destroyed, the contained objects are destroyed too.

Aggregation:
Weak "has-a" relationship.
The contained object can exist independently of the container.'''

# 18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.
# Room class
class Furniture:
    def __init__(self, name):
        self.name = name

class Appliance:
    def __init__(self, name):
        self.name = name

class Room:
    def __init__(self, name):
        self.name = name
        self.furniture = []
        self.appliances = []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

class House:
    def __init__(self, address):
        self.address = address
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def show_details(self):
        print(f"House at {self.address}:")
        for room in self.rooms:
            print(f" Room: {room.name}")
            print("  Furniture:")
            for f in room.furniture:
                print("   -", f.name)
            print("  Appliances:")
            for a in room.appliances:
                print("   -", a.name)

chair = Furniture("Chair")
table = Furniture("Table")
fridge = Appliance("Fridge")
oven = Appliance("Oven")

kitchen = Room("Kitchen")
kitchen.add_furniture(table)
kitchen.add_appliance(fridge)
kitchen.add_appliance(oven)

living_room = Room("Living Room")
living_room.add_furniture(chair)

my_house = House("123 Main St")
my_house.add_room(kitchen)
my_house.add_room(living_room)

my_house.show_details()


```

```python
# 19. How can you achieve flexibility in composed objects by allowing them to be replaced or modified
# dynamically at runtime?
'''Use setter methods or directly assign new objects to attributes, so components can be replaced at runtime.
Design your classes to depend on interfaces or base classes instead of concrete classes, allowing easy swapping.
Use dependency injection, passing new objects when needed instead of hardcoding them.
'''

# 20. Create a Python class for a social media application, using composition to represent users, posts, and
# comments.
class Comment:
    def __init__(self, user, text):
        self.user = user
        self.text = text

class Post:
    def __init__(self, content):
        self.content = content
        self.comments = []

    def add_comment(self, comment):
        self.comments.append(comment)

class User:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post):
        self.posts.append(post)

user1 = User("Alice")
post1 = Post("Hello world!")
comment1 = Comment("Bob", "Nice post!")

post1.add_comment(comment1)
user1.add_post(post1)

print(f"User: {user1.username}")
for p in user1.posts:
    print(f"Post: {p.content}")
    for c in p.comments:
        print(f"Comment by {c.user}: {c.text}")
```
